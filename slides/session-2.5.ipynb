## Why Parameterized Reporting?

- Allows you to make multiple reports at once
- Avoids copy paste issues if you were to make multiple reports by hand
- It feels like magic!

## Parameters

```
---
title: "Report on Life Expectancy"
format: html
execute: 
  echo: false
  warning: false
  message: false
---
```

. . .

```
---
title: "Report on Life Expectancy"
format: html
execute: 
  echo: false
  warning: false
  message: false
params: 
  country: Afghanistan 
---
```
